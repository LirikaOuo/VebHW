```python
# Работа с данными
import pandas as pd
import numpy as np

# Библиотеки для визуализации
import seaborn as sns
import matplotlib.pyplot as plt
```

#### *Загрузка данных:*

```python
df = pd.read_csv("spotify-2023_fixed.csv")
df.head()
```

## ***Как выпустить идеальный трек, чтобы он всем запомнился на годы?***

### Рассмотрим сначала частоту выпуска треков в зависимости от года и месяца.

Нам понадобится столбецы месяц выпуска(released_month) и год выпуска(released_year):

```python
sns.regplot(data=df, x='released_month', y='released_year')
sns.set_theme(style='darkgrid')
sns.set(rc={'figure.figsize': (15, 10)})
plt.title("В какое время года чаще выходили песни?")
plt.xlabel("Месяц")
plt.ylabel("Год")
plt.show()
```

На граффике видно, что наибольшее количество песен выходит в большинстве случаев в новогодние празники. И, согласно вектору, выпуск песен каждый год увеличивается. Самая старая песня именно новогодняя.

### Зависит ли настроение песен от года издания? 
Нам понадобятся столбец отражающий настроение песни (mode) и столбец год выпуска(released_year)
*Major* - веселые
*Minor* - грустные

```python
sns.catplot(data=df, x='released_year', y='mode')
sns.set_theme(style='darkgrid')
sns.set(rc={'figure.figsize': (15, 10)})
plt.title("Зависимость между годом выпуска и настроением")
plt.xlabel("Год")
plt.ylabel("Настроение")
plt.show()
```

На граффике видно, что веселых песен больше и они чаще выходят. Но первая песня была грустной...

### Какой ритм у песен грусных и веселых? 
Нам понадобятся столбец отражающий настроение песни (mode) и ритм (bpm)

```python
sns.catplot(data=df, x='bpm', y='mode')
sns.set_theme(style='darkgrid')
sns.set(rc={'figure.figsize': (15, 10)})
plt.title("Зависимость между ритмом и настроением")
plt.xlabel("Ритм")
plt.ylabel("Настроение")
plt.show()
```

У мажора и минора встречаются музыка всех видов ритма. Но на минимуме больше у минора, а на максимуме у мажора больше.

### Зависит ли настроение песен от месяца издания, а от года, какой у них ритм? Соеденим все таблицы.
Нам понадобятся столбец отражающий настроение песни (mode) столбец год выпуска(released_year), месяца (released_month) и ритм (bpm)

```python
sns.scatterplot(data=df, 
                x='released_month', 
                y='released_year', 
                hue='mode', 
                palette='dark:#5A9_r', 
                size='bpm')
sns.set_theme(style='whitegrid')
sns.set(rc={'figure.figsize': (15, 10)})
plt.title("Зависимость характера и ритма треков от года и месяца выпуска")
plt.xlabel("Месяц")
plt.ylabel("Год")
plt.legend(title='Настроение и ритм')
plt.show()
```

Итог: на графике видно, что самый первый трек был грустным, но ритмичным, наверное поэтому до сих пор сохранился в памяти и попал в в эту музыкальную библиотеку. Чаще всего сохраняются именно ритмичные треки. Так же, согласно графику, лучше всего сохраняются именно веселые песни.
Вывод: чтобы повысить шансы на то, чтобы новую песню запомнили надолго, она должна быть веселой, энергичной и выпуститься во время празников, чтобы больше народу имело время ее оценить.
